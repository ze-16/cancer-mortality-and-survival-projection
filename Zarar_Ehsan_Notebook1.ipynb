Code block 17 from code re-use session 1

```python


def find_outliers_IQR(Bcancer):

  q1=Bcancer.quantile(0.25)

  q3=Bcancer.quantile(0.75)

  IQR=q3-q1

  outliers = Bcancer[((Bcancer<(q1-1.75*IQR))|(Bcancer>(q3+1.75*IQR)))]

  return outliers
  #Building a function to detect any outliers in a variable of choice.
  #To investigate inconsistent data observations.
```
